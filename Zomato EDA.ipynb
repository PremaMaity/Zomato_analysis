```python
#### my fucntion


import matplotlib.pyplot as plt

def piecharts(df,column):
    values = df[column].value_counts()[:5].values
    labels = df[column].value_counts()[:5].index
        # Total % each contribute
    percentage = [] 
    for i in values:
        percen = round(i*100/df[column].shape[0],2).astype(str)
        percentage.append(percen + '%')
    df_charts = pd.DataFrame({'Columns' : labels,'Values' :values, 'Percentage' : percentage})
    #print(df_charts)
    
    
    ## define custom color palette
    custom_palette=['Orange', 'White', 'Yellow', 'Green', (0, 0.2745, 0)]
    
    ## create a color map using custom palette
    color_map = dict(zip(labels[:5], custom_palette))
    plt.figure(figsize=(6, 9))
    plt.pie(values, autopct='%1.2f%%', shadow=True, textprops={'fontsize': 10, 'color': 'black'},
            colors=[color_map[x] for x in labels])
    plt.legend(labels, fontsize=15, loc='upper left', bbox_to_anchor=(1, 0.5))
    plt.title(column, fontsize=20)
    plt.show()  
    print(df_charts)
```

```python
#### With lambda function without % sign


def mypie(df, column):
    df_charts = df[column].value_counts()[:5].reset_index()
    df_charts.columns = ['Columns', 'Percentage']
    df_charts['Percentage'] = df_charts['Percentage'].apply(lambda x:round(x*100/df[column].shape[0],2))
    plt.pie(df_charts['Percentage'], labels=df_charts['Columns'])
    plt.show()
    return df_charts
```

```python
#### With lambda function with % sign

def mycharts(df, column):
    df_charts = df[column].value_counts()[:5].reset_index()
    df_charts.columns = ['Columns', 'Count']
    df_charts['Percentage'] = df_charts['Count'] * 100 / df[column].shape[0]
    df_charts['Percentage'] = df_charts['Percentage'].apply(lambda x: '{:.2f}%'.format(x))
    print(df_charts)
    
    plt.pie(df_charts['Count'], labels=df_charts['Columns'])
    plt.show()
```

```python
#### checking all the numerical columns::

def mynum(df, column):
    df_charts = df[column].value_counts().reset_index()
    df_charts.columns = ['Columns', 'Total Count']
    df_charts['Percentage'] = df_charts['Total Count'] * 100 / df[column].shape[0]
    df_charts['Percentage'] = df_charts['Percentage'].apply(lambda x: '{:.2f}%'.format(x))
    print(df_charts[:5])
    
    ### bar plot
    plt.figure(figsize =(10,6))
    sns.barplot(x= 'Total Count', y='Columns' , data = df_charts, orient='h')
    plt.tick_params(axis='both', labelsize=9)
    plt.title(column,fontsize =15)
    plt.show()
```
